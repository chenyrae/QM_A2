```python
crime_bur = pd.read_csv('crime_burglary2011.csv')
edu2011 = pd.read_csv('education2011.csv')
age = pd.read_csv('Age.csv')
house_p = pd.read_csv('Median_houseP.csv')
BAME = pd.read_csv('BAME.csv')
household_indi = pd.read_csv('individual_household.csv')
income = pd.read_csv('median_income.csv')
car = pd.read_csv('car.csv')
household_income = pd.read_csv('household_income.csv')
population = pd.read_csv('population.csv')
```

```python
#crime_edu = crime_bur.merge(edu2011, left_on='Row_Labels', right_on='mnemonic')
```

```python
Bur_age = crime_bur.merge(age, left_on='Row_Labels', right_on='mnemonic')
```

```python
crime_BAME = Bur_age.merge(BAME, left_on='Row_Labels', right_on='LSOAs')
```

```python
crime_indi = crime_BAME.merge(household_indi, left_on='Row_Labels', right_on='LSOAs')
```

```python
crime_income = crime_indi.merge(income, left_on='Row_Labels', right_on='Code')
```

```python
crime_car = crime_income.merge(car, left_on='Row_Labels', right_on='LSOAs')
```

```python
crime_household = crime_car.merge(household_income, left_on='Row_Labels', right_on='LSOAs')
```

```python
crime_pop = crime_household.merge(population, left_on='Row_Labels', right_on='LSOA')
```
